```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
def convert_to_seconds(time_str):
    minu, sec = time_str.split("m")
    s = sec[:-1]
    seconds = float(s)
    minutes = float(minu)
    return minutes * 60 + seconds
```

```python
data = pd.read_csv('test_prod_logs.csv')
```

```python
dataproc09 = data[data.cpuname=="slurmstepd-dataproc09"]
dataproc10 = data[data.cpuname=="slurmstepd-dataproc10"]
dataproc11 = data[data.cpuname=="slurmstepd-dataproc11"]
dataproc12 = data[data.cpuname=="slurmstepd-dataproc12"]
```

```python
Ngen_L = 1e4
Ngen_M = 1e3
Ngen_H = 1e2
Ngen_UH = 10

dataproc09L = dataproc09[dataproc09.minimal_energy == 1e3]
dataproc10L = dataproc10[dataproc10.minimal_energy == 1e3]
dataproc11L = dataproc11[dataproc11.minimal_energy == 1e3]
dataproc12L = dataproc12[dataproc12.minimal_energy == 1e3]

dataproc09M = dataproc09[dataproc09.minimal_energy == 1e5]
dataproc10M = dataproc10[dataproc10.minimal_energy == 1e5]
dataproc11M = dataproc11[dataproc11.minimal_energy == 1e5]
dataproc12M = dataproc12[dataproc12.minimal_energy == 1e5]

dataproc09H = dataproc09[dataproc09.minimal_energy == 1e7]
dataproc10H = dataproc10[dataproc10.minimal_energy == 1e7]
dataproc11H = dataproc11[dataproc11.minimal_energy == 1e7]
dataproc12H = dataproc12[dataproc12.minimal_energy == 1e7]

dataproc09UH = dataproc09[dataproc09.minimal_energy == 1e8]
dataproc10UH = dataproc10[dataproc10.minimal_energy == 1e8]
dataproc11UH = dataproc11[dataproc11.minimal_energy == 1e8]
dataproc12UH = dataproc12[dataproc12.minimal_energy == 1e8]
```

```python
dataproc09L["real_time_seconds"] = dataproc09L["real_time"].apply(convert_to_seconds)
dataproc10L["real_time_seconds"] = dataproc10L["real_time"].apply(convert_to_seconds)
dataproc11L["real_time_seconds"] = dataproc11L["real_time"].apply(convert_to_seconds)
dataproc12L["real_time_seconds"] = dataproc12L["real_time"].apply(convert_to_seconds)

dataproc09M["real_time_seconds"] = dataproc09M["real_time"].apply(convert_to_seconds)
dataproc10M["real_time_seconds"] = dataproc10M["real_time"].apply(convert_to_seconds)
dataproc11M["real_time_seconds"] = dataproc11M["real_time"].apply(convert_to_seconds)
dataproc12M["real_time_seconds"] = dataproc12M["real_time"].apply(convert_to_seconds)

dataproc09H["real_time_seconds"] = dataproc09H["real_time"].apply(convert_to_seconds)
dataproc10H["real_time_seconds"] = dataproc10H["real_time"].apply(convert_to_seconds)
dataproc11H["real_time_seconds"] = dataproc11H["real_time"].apply(convert_to_seconds)
dataproc12H["real_time_seconds"] = dataproc12H["real_time"].apply(convert_to_seconds)

dataproc09UH["real_time_seconds"] = dataproc09UH["real_time"].apply(convert_to_seconds)
dataproc10UH["real_time_seconds"] = dataproc10UH["real_time"].apply(convert_to_seconds)
dataproc11UH["real_time_seconds"] = dataproc11UH["real_time"].apply(convert_to_seconds)
dataproc12UH["real_time_seconds"] = dataproc12UH["real_time"].apply(convert_to_seconds)
```

```python
binst = np.logspace(-1 , 4.1)
```

```python
plt.hist(dataproc09UH["real_time_seconds"]/Ngen_UH, alpha = 0.5, label = 'UH', bins = binst)
plt.hist(dataproc09H["real_time_seconds"]/Ngen_H, alpha = 0.5, label = 'H', bins = binst)
plt.hist(dataproc09M["real_time_seconds"]/Ngen_M, alpha = 0.5, label = 'M', bins = binst)
plt.hist(dataproc09L["real_time_seconds"]/Ngen_L, alpha = 0.5, label = 'L', bins = binst)
plt.xscale("log")
plt.xlabel("real time per event (s)")
plt.ylabel("n jobs")
plt.title("dataproc09")
plt.legend()
```
