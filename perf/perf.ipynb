```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
def convert_to_seconds(time_str):
    minu, sec = time_str.split("m")
    s = sec[:-1]
    seconds = float(s)
    minutes = float(minu)
    return minutes * 60 + seconds
```

```python
def convert_to_seconds(time_str):
    if pd.isnull(time_str):  # Controllo per valori mancanti
        return None
    if isinstance(time_str, (int, float)):  # Controllo per valori numerici
        return float(time_str)  # Lascia i numeri così come sono
    try:
        # Separare minuti e secondi
        minutes, seconds = time_str.split('m')
        seconds = seconds.replace('s', '')  # Rimuovere la 's'
        
        # Convertire minuti e secondi in float
        total_seconds = int(minutes) * 60 + float(seconds)
        return total_seconds
    except Exception as e:
        print(f"Errore nella conversione di: {time_str} -> {e}")
        return None  # Restituisci None se il formato non è valido
```
